```python
from datetime import datetime
```

```python
class Pedido:

  def __init__(self, id_producto, cant_producto, precio_pedido, direccion_entrega, metodo_pago):
    '''
    Método constructor de la clase. Recibe los argumentos correspondientes a la información básica del pedido.
    '''
    self.id_producto = id_producto
    self.cant_producto = cant_producto
    self.precio_pedido = precio_pedido
    self.direccion_entrega = direccion_entrega
    self.metodo_pago = metodo_pago
    self.estado = "Por entregar" #Estado por defecto al momento de crear el producto
    self.hora_generacion = datetime.now() #Marca con la hora de generación del pedido

  def ticket_pedido(self):
    '''
    Método para imprimir el ticket del pedido con toda la información que necesita el domiciliario.
    No recibe ningún argumento.
    '''
    print('Informacón del pedido a entregar: ')
    print(f'Cantidad de productos a entregar: {self.cant_producto}')
    print(f'Precio del pedido: {self.precio_pedido}')
    print(f'Dirección de entrega: {self.direccion_entrega}')
    print(f'Método de pago: {self.metodo_pago}')
    print(f'Estado: {self.estado}')
    print(f'Hora de generación del pedido: {self.hora_generacion}')

  def cambiar_estado(self):
    """
    Función que permite cambiar el estado del pedido. No pide ningún argumento de entrada. Posterior a su llamado pide un dígito
    de acuerdo al cambio que se quiere realizar.

    1: Para cambiar el estado a "Entregado"
    2: Pra cambiar el estado a "Cancelado"
    3: Para abortar el cambio de estado.
    """
    opcion = int(input("Digite 1 si el pedido fue entregado, 2 si el pedido fue cancelado, 3 si desea regresar el estado a Por entregar o 0 si desea cancelar esta acción: "))

    if opcion == 1:
      self.estado = "Entregado"
      print(f'El pedido se ha entregado en la dirección {self.direccion_entrega}')
    elif opcion == 2:
      self.estado = "Cancelado"
      print('El pedido se ha cancelado')
    elif opcion == 3:
      self.estado = "Por entregar"
      print('El pedido ha vuelto al estado "Por entregar"')
    elif opcion == 0:
      print('Accion cancelada')
    else:
      print("El dígito ingresado es incorrecto, intente nuevamente.")

  def prioridad(self):
    '''
    Función que permite consultar la prioridad de entrega del pedido de acuerdo a las políticas de la empresa.
    La función determina cuantos minutos han pasado desde la generación del pedido hasta la consulta de la prioridad.
    <= 15 minutos: A tiempo
    > 15 minutos y <= 30: Momento de entregar el pedido
    > 30: Entrega urgente!
    '''

    #Validación inicial del estado del producto

    if self.estado == "Entregado":
      print("El pedido ya fue entregado.")
    elif self.estado == "Cancelado":
      print("El pedido fue cancelado.")

    else:
      hora_consulta_prioridad = datetime.now() #Generación de la marca de hora del momento de consulta
      diferencia = (hora_consulta_prioridad - self.hora_generacion).total_seconds() / 60 #Cálculo de los minutos transcurridos

      if diferencia <= 15:
        print(f"Pedido generado hace {diferencia:.2f} minutos. Aún está a tiempo.")
      elif diferencia > 15 and diferencia <= 30:
        print(f"Pedido generado hace {diferencia:.2f} minutos. Es momento de entregar el producto.")
      else:
        print(f"EL PEDIDO SE GENERÓO HACE {diferencia:.2f} MINUTOS. TIEMPO INCUMPLIDO, ENTREGUE INMEDIATAMENTE")
```

**Utilidad de la clase:** Esta clase proporciona una estructura organizada para gestionar la información, el estado y la prioridad de despacho de los pedidos en el negocio, contribuyendo a une a gestión más eficiente y efectiva de las operaciones de venta y entrega

```python
def calcular_e(aproximaciones=1):
    """
    Calcula una aproximación de la constante de Euler.
    La función recibe como argumento la cantidad de series a utilizar para realizar las aproximaciones.
    Return: Constante de Euler aproximada
    """
    e_aproximado = 0
    factorial = 1

    for i in range(aproximaciones):
        e_aproximado += 1 / factorial
        factorial *= (i + 1)

    return e_aproximado
```

```python
#Cantidad de aproximaciones necesarias
calcular_e(10)
```
