## **Ejercicio 1** (40%)

Construya un DataFrame a partir del siguiente diccionario

`exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}`

con las siguientes etiquetas para los índices

`labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']`

1. Seleccione solo las columnas *name* y *score*.
  
2. Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

4. Cambie el *score* en la fila 'd' a 11.5

5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

7. Ordene el DataFrame por nombre en orden ascendente.

8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

```python
import pandas as pd
import numpy as np
```

Ejercicio 1

```python
# Definimos el diccionario
data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df.head()
```

```python
# Seleccionamos las columnas name y score
name_score = df[['name','score']]
name_score
```

```python
# Numero de intentos >2
filtro_attempts = df[df['attempts'] > 2]
filtro_attempts
```

```python
# Score entre 15 y 20
filtro_score = df[(df['score']>= 15) & (df['score']<=20)]
filtro_score
```

```python
# Cambiar el score de la fila d a 11.5
df.loc['d', 'score']= 11.5
df
```

```python
# Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna
nuevos_valores = {'name': 'Julia' , 'score': 22.0, 'attempts': 1, 'qualify': 'yes'}
df = df.append(pd.Series(nuevos_valores, name = 'k'  ))
df
```

```python
#Eliminar fila K
df = df.drop('k')
df
```

```python
# Ordene el DataFrame por nombre en orden ascendente
df_ordenado = df.sort_values(by='name',ascending=True)
df_ordenado
```

```python
# Reemplace los valores de la columna qualify que contiene los valores yes y no por los valores booleanos True y False respectivamente.
df['qualify'] = df['qualify'].apply(lambda x: True if x == 'yes' else False)
df
```

## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

```python
athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")
athletes.head()
```

```python
# Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
medallas= athletes[(athletes['gold']>= 1) & (athletes['silver']<=1)&(athletes['bronze']>= 1)]
medallas.head()
```

```python
# Filtre aplicado a columna calculada
medallas.loc[:,'cant_medallas']=medallas.loc[:,'gold']+medallas.loc[:,'silver']+medallas.loc[:,'bronze']
medallas.loc[medallas.loc[:,'cant_medallas']>0,:].head()
```

```python
#Calcule la masa muscular por fila con una función como
athletes.loc[:,'imc']=athletes.loc[:,'weight']/(athletes.loc[:,'height']**2)
athletes.head()
```

```python
# Agrupe los datos calculando la media por deporte para cada una del resto de características.

pd.DataFrame(athletes.groupby('sport').imc.mean())
```

```python
# ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?
sports_imc= pd.DataFrame(athletes.groupby('sport').imc.mean())
sports_imc.reset_index(inplace=True)
sports_imc.loc[sports_imc['imc'].idxmax(), ['sport','imc']]
```

```python
# ¿Cuál es el pais con más participación femenina?
(athletes[athletes['sex'] == 'female'])
(athletes.sort_values(by='sex', ascending=True))
```
